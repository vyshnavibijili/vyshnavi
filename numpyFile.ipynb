```python
val = [("Popo",2007,8.6),("Dodo",2009,6.7),("Lolo",2004,9.2)]
dtype1 = [("name",'S10'),("year",int),("gpa",float)]
arr = np.array(val,dtype=dtype1)
print(arr)

print(np.sort(arr,order="name"))
print(np.sort(arr,order="gpa"))
print(np.sort(arr,order=["year","gpa"]))
```
